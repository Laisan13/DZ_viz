```python
dynamics_happiness = df.groupby('Overall rank').sum()
dynamics_happiness
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
dynamics_happiness['Score'].plot(title='Рейтинг стран мира по уровню счастья', 
                                        xlabel='Рейтинг стран', 
                                        ylabel='Оценка счастья, баллы')
"""Чем выше рейтинг, тем ниже по ступени страна находится и тем ниже оценка счастья в данной стране"""
```

```python
ax = dynamics_happiness.plot(title='Рейтинг уровня счастья от показателей благополучия', 
                             xlabel='Рейтинг стран', 
                             ylabel='Показатели благополучия')
ax.legend(loc='upper left')
ax.text(0, 2, 'Здесь линии более ровные')
ax.text(130, 2, 'Здесь сильные скачки') 
"""Прослеживается конечно некоторое снижение показателей в зависимости от рейтинга страны. 
Судя по "ровности" линий, чем ниже рейтинг страны тем скачкоообразнее показатели, как в детекторе лжи.
Мое мнение, что показатели подтасованы, у стран где сильные скачки"""
```

```python
# Area chart
ax_1 = dynamics_happiness.plot(kind='area',
                             stacked=False,
                             title='Рейтинг уровня счастья от показателей благополучия', 
                             xlabel='Рейтинг стран', 
                             ylabel='Показатели благополучия')
"""Не совсем удачная визуализация для принятых данных, но здесь видно так же скачки у показателей, и показатель """
```

```python
# Гистограмма
"""Из гистограммы видно что больше всего оценок попало в интервал от 5 до 5,5 , и ниже 2,8 оценок нет совсем. 
Так как из прошлого графика было понятно чем выше оценка счастья тем выше по ступени развития и благополучия стоит страна.
Из гистограммы видно что стран с низкой оценкой от 2,8 до 4,3 , примерно столько же сколько стран с высокой оценкой от 6,5 до 8.
Некое равновесие счастья что ли?!)))"""
dynamics_happiness['Score'].plot(kind='hist',
                   bins=80,
                   title='Рейтинг счастья',
                   xlabel='Оценка счастья',
                   ylabel='Частота попадания оценки в каждый интервал',
                   xticks=range(8))

plt.xlabel('Оценка счастья')
plt.ylabel('Частота попадания оценки в каждый интервал')
```
